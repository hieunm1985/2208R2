```python
'''
Bài 14: Nhập vào một chuỗi s chỉ chứa các ký tự '(', ')', '{', '}', '[' và ']', 
xác định xem chuỗi s có hợp lệ hay không.

Chuỗi đầu vào hợp lệ nếu:

- Các dấu ngoặc mở phải được đóng bằng cùng một loại dấu ngoặc.
- Dấu ngoặc mở phải được đóng theo đúng thứ tự.
- Mọi dấu ngoặc đóng đều có một dấu ngoặc mở tương ứng cùng loại.

Ví dụ 1:
Input: s = "()"
Output: True

Ví dụ 2:
Input: s = "() [] {}"
Output: True

Ví dụ 3:
Input: s = "(]"
Output: False
'''
```

```python
str = input("nhap vao 1 chuoi: ")
check = ""
if len(str) % 2 != 0:
    check = "chuoi khong hop le"
else:
    for i in range(0,len(str),2):
        if str[i] + str[i+1] == "()" or str[i] + str[i+1] == "{}" or str[i] + str[i+1] == "[]":
            check = "chuoi hop le"
        else:
            check = "chuoi khong hop le"
            break
print(check)
```

```python
'''
Bài 15: Roman to Integer
Chữ số La Mã được thể hiện bằng bảy ký hiệu khác nhau: I, V, X, L, C, D và M.

Symbol       Value
I             1
V             5
X             10
L             50
C             100
D             500
M             1000

Ví dụ: 2 được viết là II bằng số La Mã, chỉ là hai cái được cộng lại với nhau. 
12 được viết là XII, đơn giản là X + II. Con số 27 được viết là XXVII, là XX + V + II.

Chữ số La mã thường được viết từ lớn nhất đến nhỏ nhất từ trái sang phải. 
Tuy nhiên, chữ số cho bốn không phải là IIII. Thay vào đó, số bốn được viết là IV. 
Bởi vì cái đứng trước năm, chúng ta trừ nó ra làm bốn. Nguyên tắc tương tự cũng áp dụng cho số chín, 
được viết là IX. Có sáu trường hợp phép trừ được sử dụng:

I có thể được đặt trước V (5) và X (10) để tạo thành 4 và 9.
X có thể được đặt trước L (50) và C (100) để tạo ra 40 và 90.
C có thể được đặt trước D (500) và M (1000) để tạo thành 400 và 900.
Nhập vào một chuỗi đại diện cho một số La Mã, hãy chuyển nó thành một số nguyên.

Ví dụ 1:
Input: s = "III"
Output: 3
Explanation: III = 3.

Ví dụ 2:
Input: s = "LVIII"
Output: 58
Explanation: L = 50, V= 5, III = 3.

Ví dụ 3:
Input: s = "MCMXCIV"
Output: 1994
Explanation: M = 1000, CM = 900, XC = 90 and IV = 4.
'''
```

```python
str = input("nhap vao chuoi: ")
n = 0
i = 0
if len(str) == 1:
    if str[len(str)-1] == "I":
        n = n + 1
    elif str[len(str)-1] == "V":
        n = n + 5
    elif str[len(str)-1] == "X":
        n = n + 10
    elif str[len(str)-1] == "L":
        n = n + 50
    elif str[len(str)-1] == "C":
        n = n + 100
    elif str[len(str)-1] == "D":
        n = n + 500
    elif str[len(str)-1] == "M":
        n = n + 1000 
else:
    while i <= (len(str) - 2):
        if str[i] + str[i+1] == "IV":
            n = n + 4
            i += 2
        elif str[i] + str[i+1] == "IX":
            n = n + 9
            i += 2
        elif str[i] + str[i+1] == "XL":
            n = n + 40
            i += 2
        elif str[i] + str[i+1] == "XC":
            n = n + 90
            i += 2
        elif str[i] + str[i+1] == "CD":
            n = n + 400
            i += 2
        elif str[i] + str[i+1] == "CM":
            n = n + 900
            i += 2
        else:
            if str[i] == "I":
                n = n + 1
                i += 1
            elif str[i] == "V":
                n = n + 5
                i += 1
            elif str[i] == "X":
                n = n + 10
                i += 1
            elif str[i] == "L":
                n = n + 50
                i += 1
            elif str[i] == "C":
                n = n + 100
                i += 1
            elif str[i] == "D":
                n = n + 500
                i += 1
            elif str[i] == "M":
                n = n + 1000
                i += 1
    if str[len(str)-3] + str[len(str)-2] in ["IV","IX","XL","XC","CD","CM"] or str[len(str)-2] + str[len(str)-1] not in ["IV","IX","XL","XC","CD","CM"]:
        if str[len(str)-1] == "I":
            n = n + 1
        elif str[len(str)-1] == "V":
            n = n + 5
        elif str[len(str)-1] == "X":
            n = n + 10
        elif str[len(str)-1] == "L":
            n = n + 50
        elif str[len(str)-1] == "C":
            n = n + 100
        elif str[len(str)-1] == "D":
            n = n + 500
        elif str[len(str)-1] == "M":
            n = n + 1000

print(f'so la ma {str} co gia tri la {n}')
```

```python
string = input()
s = 0
for i in range(0,len(string)):
    if string[i] == 'I':
        s +=1
    elif string[i] == 'V':
        if string[i-1] == 'I':
            s+=4-1
        else:
            s+=5
    elif string[i] == 'X':
        if string[i-1] == 'I':
            s+=9-1
        else:
            s+=10
    elif string[i] == 'L':
        if string[i-1] == 'X':
            s+=40-10
        else:
            s+=50
    elif string[i] == 'C':
        if string[i-1] == 'X':
            s+=90-10
        else:
            s+=100
    elif string[i] == 'D':
        if string[i-1] == 'C':
            s+=400-100
        else:
            s+=500
    elif string[i] == 'M':
        if string[i-1] == 'C':
            s+=900-100
        else:
            s+=1000
print(s)
```

```python
# vẽ hình vuông:
import turtle
pen = turtle.Turtle()
mau = ["blue", 'red', 'pink', 'green']
for i in range(4):
    pen.pencolor(mau[i])
    pen.fd(100)
    pen.left(90)
turtle.done()
```

```python
# vẽ ngôi sao:
import turtle
pen = turtle.Turtle()
pen.speed(50)
pen.left(69)
mau = ['red', 'blue', 'orange', 'green', 'pink']
for j in range(80):
    pen.pencolor(mau[j%5])
    pen.up()
    pen.circle(50,5)
    pen.down()
    for i in range(5):
        pen.fd(100)
        pen.rt(144)
turtle.done()
```
