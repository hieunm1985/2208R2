##bài số 3
nhập vào tiền lương theo giờ.
số giờ làm việc trong 1 tháng.
nhập tienf thưởng.
in ra tổng thu nhập trong tháng biết tỏng thu nhập = lương*số giờ + tiền thưởng.

```python
tien_luong = int(input("nhập tiền lương theo giờ: "))
so_gio = float(input("nhập số giờ làm việc trong tháng: "))
tien_thuong = int(input("nhập tiền thưởng: "))
tong_thu_nhap = tien_luong * so_gio + tien_thuong
print(f" tổng thu nhập là {tong_thu_nhap}")
```

Bài số 4
nhập vào 1 string và 1 số x. hãy xóa x ký tự đầu tiên của string.


```python
name = input("nhập vào chuỗi: ")
x = int(input("nhập vào số ký tự cần xóa: "))
print(f"văn bản mới là: {name[x:]}")
```

bài số 5:
nhập vào 1 string và kiểm tra xem đây có phải là 1 string đối xứng hay không? Nếu đối xứng in ra true, không thì false.
Chuỗi đối xứng là chuỗi có thứ tự giống nhau khi đọc từ trái qua phải và ngược lại.
ví dụ: abba, abcdcba

```python
name1 = input("nhập vào chuỗi: ")
y = len(name1)//2
z = name1[0:y:1] == name1[-1:-(y+1):-1]
print(z)
```
